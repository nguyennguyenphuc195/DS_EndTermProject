```python
def get_id(track_name: str, token: str) -> str:
    headers = {
        'Accept': 'application/json',
        'Content-Type': 'application/json',
        'Authorization': f'Bearer ' + token,
    }
    params = [
        ('q', track_name),
        ('type', 'track'),
    ]
    try:
        response = requests.get('https://api.spotify.com/v1/search', 
                    headers = headers, params = params, timeout = 5)
        json = response.json()
        first_result = json['tracks']['items'][0]
        track_id = first_result['id']
        return track_id
    except:
        return None
```

```python
def get_features(track_id: str, token: str) -> dict:
    headers = {
        'Accept': 'application/json',\
        'Content-Type': 'application/json',\
        'Authorization': f'Bearer ' + token,\
    }

    response = requests.get(f"https://api.spotify.com/v1/audio-features/{track_id}", headers = headers)
    json_obj = response.json()
    
    if response.status_code != 200:
        print(response)
        return response, None
    
    return response, json_obj
```

# Lấy danh sách các playlist do spotify tổng hợp

```python
def get_all_playlist(username: str, token: str) -> List[Dict]:
    headers = {
        'Accept': 'application/json',\
        'Content-Type': 'application/json',\
        'Authorization': f'Bearer ' + token,\
    }
    
    offset   = 0
    limit    = 50
    total    = 100000 #max offset in spotify
    playlist = []
    
    while offset < total:
        params = [("limit", limit), ("offset", offset)]
        
        response = requests.get(f"https://api.spotify.com/v1/users/{username}/playlists", headers = headers, params=params)
        json_obj = response.json()
        
        if response.status_code != 200:
            print(response)
            continue
        
        playlist.extend(json_obj["items"])
        n = offset + len(json_obj["items"])

        print(f"Successfully get from playlist {offset} to {n}")
        offset = n
        
        if json_obj["total"] < total:
            total = json_obj["total"]
            
        if not "next" in json_obj or json_obj["next"] == "":
            break

    return playlist
```

# Lấy tất cả các dữ liệu về track

```python
def get_several_tracks(track_ids: str, token: str) -> dict:
    headers = {
        'Accept': 'application/json',\
        'Content-Type': 'application/json',\
        'Authorization': f'Bearer ' + token,\
    }

    params   = [('ids', ",".join(track_ids))]
    response = requests.get(f"https://api.spotify.com/v1/tracks", headers = headers, params=params)
    json_obj = response.json()
    
    if response.status_code != 200:
        print(response)
        return response, None
    
    return response, json_obj
```

```python
def get_features_several_tracks(track_ids: str, token: str) -> dict:
    headers = {
        'Accept': 'application/json',\
        'Content-Type': 'application/json',\
        'Authorization': f'Bearer ' + token,\
    }

    params   = [('ids', ",".join(track_ids))]
    response = requests.get(f"https://api.spotify.com/v1/audio-features", headers = headers, params=params)
    json_obj = response.json()
    
    if response.status_code != 200:
        print(response)
        return response, None
    
    return response, json_obj
```

### Get tracks

```python
def get_tracks(token: str):
    tracks = []
    i   = 0
    ids = []
    with open("tracks.txt") as f:
        for line in f:
            ids.append(line[:-1])
            i += 1
            if i % 50 == 0:
                res, json_obj = get_several_tracks(ids, token)
                new_tracks = []
                if res.status_code == 200:
                    new_tracks = json_obj["tracks"]
                elif res.status_code == 401:
                    token = renew_token()
                    res, json_obj = get_several_tracks(ids, token)
                    if res.status_code == 200:
                        new_tracks = json_obj["tracks"]
                tracks.extend(new_tracks)
                ids = []
                print("Get", len(new_tracks), "tracks")

    return tracks
```

```python
tracks = get_tracks(token)
```

```python
def get_row_from_track(track: str):
    #Tạo dòng
    row = {}
    
    row["id"]      = track["id"]
    row["name"]    = track["name"]
    row["artists"] = ",".join([a["id"] for a in track["artists"]])
    row["available_markets"] = ",".join(track["available_markets"])
    row["explicit"]   = track["explicit"]
    row["popularity"] = track["popularity"]
    return row
```

# Lấy tất cả các dữ liệu về artist

```python
def get_several_artist(artist_ids: str, token: str) -> dict:
    headers = {
        'Accept': 'application/json',\
        'Content-Type': 'application/json',\
        'Authorization': f'Bearer ' + token,\
    }

    params   = [('ids', ",".join(artist_ids))]
    response = requests.get(f"https://api.spotify.com/v1/artists", headers = headers, params=params)
    json_obj = response.json()
    
    if response.status_code != 200:
        print(response)
        return response, None
    
    return response, json_obj
```

```python
def get_row_from_artist(artist):
    #Tạo dòng
    row = {}
    row["followers"] = artist["followers"]["total"]
    row['genres'] = artist['genres']
    row['id'] = artist['id']
    row['name'] = artist['name']
    row['popularity'] = artist['popularity']

    return row
```

```python
def get_artists(token: str):
    tracks = []
    i   = 0
    ids = []
    new_tracks = []
    with open("artists.txt") as f:
        for line in f:
            ids.append(line[:-1])
            i += 1
            if i % 50 == 0:
                res, json_obj = get_several_artist(ids, token)
                new_tracks = []
                if res.status_code == 200:
                    if "artists" in json_obj:
                        new_tracks = json_obj["artists"]
                    else:
                        print(json_obj)
                elif res.status_code == 401:
                    token = renew_token()
                    res, json_obj = get_several_tracks(ids, token)
                    if res.status_code == 200:
                        if "artists" in json_obj:
                            new_tracks = json_obj["artists"]
                tracks.extend(new_tracks)
                ids = []
                
            if i % 10000 == 0:
                print("Get 10000 artists")
    
    if len(ids) != 0:
        res, json_obj = get_features_several_tracks(ids, token)
        new_tracks = []
        if res.status_code == 200:
            if "artists" in json_obj:
                new_tracks = json_obj["artists"]
        elif res.status_code == 401:
            token = renew_token()
            res, json_obj = get_several_tracks(ids, token)
            if res.status_code == 200:
                if "artists" in json_obj:
                    new_tracks = json_obj["artists"]
        tracks.extend(new_tracks)
        
    return tracks
```
